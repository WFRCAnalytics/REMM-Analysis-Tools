# final edits on pivoted control totals

```python
# unpivot (stack) pivoted data from excel
pop_controls = pd.read_csv(r"E:\REMM\REMMRun_Max_Sprawl\population_controls_pivoted.csv", index_col=0)
pop_controls_stack = pop_controls.stack()
pop_controls_stack = pop_controls_stack.reset_index()
pop_controls_stack.columns = ['year', 'cid', 'number_of_population']

# # truncate decimals so control total is not exceeded and export
pop_controls_stack[['year','number_of_population','cid']].astype(int).to_csv(r".\Outputs\population_controls_maxsprawl.csv", index=False)
pop_controls_stack.head(20)
```

```python
# unpivot (stack) pivoted data from excel
hh_controls = pd.read_csv(r"E:\REMM\REMMRun_Max_Sprawl\household_controls_pivoted.csv", index_col=0)
hh_controls_stack = hh_controls.stack()
hh_controls_stack = hh_controls_stack.reset_index()
hh_controls_stack.columns = ['year', 'income_quartile', 'number_households']

# forward fill years
hh_controls_stack['year']=pop_controls_stack['year'].fillna(method='ffill')


# fix income quartiles and cid (co_fips)
from itertools import cycle
cids = cycle([11,11,11,11,35,35,35,35,49,49,49,49,57,57,57,57])
iqs = cycle([1,2,3,4])
hh_controls_stack['cid'] = [next(cids) for cid in range(len(hh_controls_stack))]
hh_controls_stack['income_quartile'] = [next(iqs) for iq in range(len(hh_controls_stack))]

# truncate decimals so control total is not exceeded and export
hh_controls_stack['year','number_households','cid','income_quartile'].astype(int).to_csv(r".\Outputs\household_controls_maxsprawl.csv", index=False)
hh_controls_stack.head(20)
```
