```python
def read_se_file_for_year(path, version, year):
    if version == '832':
        csv = os.path.join(path, f'SE_{year}.csv')
    if version == '900':
        csv = os.path.join(path, f'taz900_SE_{year}.csv')
    
    df = pd.read_csv(csv)
    cols = [c for c in list(df.columns) if c != ';TAZID']

    to_replace = [c for c in list(df.columns) if not c in (';TAZID',  'CO_TAZID', 'CO_FIPS',	'CO_NAME')]
    new_col_names = [c + '_' + os.path.basename(path) for c in to_replace]
    replace_dict = dict(zip(to_replace, new_col_names))
    df.rename(replace_dict, axis=1, inplace=True)
    return df

```

```python
def calculate_stats(read_se_output):
    df_combined = pd.concat(read_se_output, axis=1)
    df_combined = df_combined.loc[:,~df_combined.columns.duplicated()].copy()

    data_cols = ['TOTHH','HHPOP', 'HHSIZE', 'TOTEMP', 'RETEMP', 'INDEMP', 'OTHEMP', 'ALLEMP', 'RETL', 'FOOD', 'MANU', 'WSLE', 'OFFI', 'GVED', 'HLTH', 'OTHR', 'FM_AGRI', 'FM_MING', 'FM_CONS', 'HBJ']
    
    for data_col in data_cols:
        cols = [col for col in df_combined.columns if data_col in col]
        df_combined[data_col + '_mean'] = df_combined.loc[:, cols].mean(axis = 1)
        df_combined[data_col + '_min'] = df_combined.loc[:, cols].min(axis = 1) # med, std, min, max, var
        df_combined[data_col + '_max'] = df_combined.loc[:, cols].max(axis = 1)
        df_combined[data_col + '_range'] = df_combined[data_col + '_max'] - df_combined[data_col + '_min']

    return df_combined
```

```python
df_with_stats = calculate_stats(dfs_900_2019)
```
